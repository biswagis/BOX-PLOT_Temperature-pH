```python
#TEMPERATURE BOX PLOT
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# -----------------------------
# Step 1: Data
# -----------------------------
data = {
    30: [42, 44, 45, 47, 48],
    40: [50, 52, 53, 55, 56],
    50: [60, 62, 63, 65, 66],
    60: [68, 69, 70, 72, 73]
}

df = pd.DataFrame(dict([(k, pd.Series(v)) for k, v in data.items()]))

# -----------------------------
# Step 2: Style
# -----------------------------
sns.set(style="white")  # remove grid

plt.rcParams['font.family'] = 'Times New Roman'
plt.rcParams['font.size'] = 14

# -----------------------------
# Step 3: Plot
# -----------------------------
fig, ax = plt.subplots(figsize=(8,6))

colors = ['#FF6F61', '#6B5B95', '#88B04B', '#F7CAC9']

box = ax.boxplot(df.values,
                 patch_artist=True,
                 widths=0.6,
                 showmeans=True,
                 meanline=True)

# Apply colors
for patch, color in zip(box['boxes'], colors):
    patch.set_facecolor(color)
    patch.set_edgecolor('black')
    patch.set_linewidth(1.5)

# Style elements
for median in box['medians']:
    median.set(color='black', linewidth=2)

for whisker in box['whiskers']:
    whisker.set(color='black', linewidth=1.5)

for cap in box['caps']:
    cap.set(color='black', linewidth=1.5)

for mean in box['means']:
    mean.set(color='blue', linewidth=2)

# -----------------------------
# Step 4: Labels
# -----------------------------
ax.set_xticks([1,2,3,4])
ax.set_xticklabels(['30', '40', '50', '60'], fontweight='bold')

ax.set_xlabel('Temperature variation', fontsize=16, fontweight='bold')
ax.set_ylabel('Adsorption (%)', fontsize=16, fontweight='bold')

# Make Y-axis numbers bold
for label in ax.get_yticklabels():
    label.set_fontweight('bold')

# -----------------------------
# Step 5: Only outer box (spines)
# -----------------------------
for spine in ax.spines.values():
    spine.set_visible(True)
    spine.set_linewidth(1.5)
    spine.set_color('black')

# Remove inner grid completely
ax.grid(True)

# -----------------------------
# Step 6: Final layout
# -----------------------------
plt.tight_layout()

# Save high quality
plt.savefig("Temperature_Boxplot.jpeg", dpi=600, bbox_inches='tight')

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# -----------------------------
# Step 1: pH Data
# -----------------------------
data = {
    1: [12, 14, 15, 17, 18],
    3: [20, 22, 23, 24, 25],
    5: [30, 32, 33, 34, 35],
    7: [60, 65, 66, 68, 70],
    9: [40, 42, 43, 44, 45]
}

df = pd.DataFrame(dict([(k, pd.Series(v)) for k, v in data.items()]))

# -----------------------------
# Step 2: Style
# -----------------------------
sns.set(style="white")

plt.rcParams['font.family'] = 'Times New Roman'
plt.rcParams['font.size'] = 14

# -----------------------------
# Step 3: Plot
# -----------------------------
fig, ax = plt.subplots(figsize=(8,6))

# New color palette (different from temperature)
colors = ['#4E79A7', '#F28E2B', '#59A14F', '#E15759', '#B07AA1']  
# dark green, orange, dark red, dark slate gray, purple

box = ax.boxplot(df.values,
                 patch_artist=True,
                 widths=0.6,
                 showmeans=True,
                 meanline=True)

# Apply colors
for patch, color in zip(box['boxes'], colors):
    patch.set_facecolor(color)
    patch.set_edgecolor('black')
    patch.set_linewidth(1.5)

# Style elements
for median in box['medians']:
    median.set(color='black', linewidth=2)

for whisker in box['whiskers']:
    whisker.set(color='black', linewidth=1.5)

for cap in box['caps']:
    cap.set(color='black', linewidth=1.5)

for mean in box['means']:
    mean.set(color='blue', linewidth=2)

# -----------------------------
# Step 4: Labels
# -----------------------------
ax.set_xticks([1,2,3,4,5])
ax.set_xticklabels(['1', '3', '5', '7', '9'], fontweight='bold')

ax.set_xlabel('pH variation', fontsize=16, fontweight='bold')
ax.set_ylabel('Adsorption (%)', fontsize=16, fontweight='bold')

# Bold Y-axis numbers
for label in ax.get_yticklabels():
    label.set_fontweight('bold')

# -----------------------------
# Step 5: Outer box only
# -----------------------------
for spine in ax.spines.values():
    spine.set_visible(True)
    spine.set_linewidth(1.5)
    spine.set_color('black')

ax.grid(True)

# -----------------------------
# Step 6: Save + Show
# -----------------------------
plt.tight_layout()

plt.savefig("pH_boxplot.jpeg", dpi=600, bbox_inches='tight')

plt.show()
```
